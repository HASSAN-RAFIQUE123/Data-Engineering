```python
# team player name with group by team
import pprint
names=collection.aggregate([{'$group':{'_id':'$Team','names_player':{'$push':'$Name'}}}])
for name in names:
  pprint.pprint(name)
```

```python
# all record using Root
names=collection.aggregate([{'$group':{'_id':'$Team','allrecord':{'$push':'$$ROOT'}}}])
for name in names:
  pprint.pprint(name)
```

```python
# count name of player who are in COL team
names=collection.find({'Team':'COL'},{'Name':1,'_id':0})
for name in names:
  pprint.pprint(name)
```

```python
# count  of player in COL team
names=collection.count_documents({'Team':'COL'})
print(names)
```

```python
# change the name of Cortina Innovation Network to CIN
collection.update_many({'Team':'Cortina Innovation Network'},{'$set':{'Team':'CIN'}})
```

```python
# total player  in  Teams
names=collection.aggregate([{'$group':{'_id':'$Team','Total':{'$sum':1}}}])
for name in names:
  pprint.pprint(name)
```

```python
# Max number of player team
names=collection.aggregate([{'$group':{'_id':'$Team','Total':{'$sum':1}}},{'$sort':{'Total':-1}},{'$limit':1}])
for name in names:
  pprint.pprint(name)
```

```python
# Minimum number of player team
names=collection.aggregate([{'$group':{'_id':'$Team','Total':{'$sum':1}}},{'$sort':{'Total':1}},{'$limit':1}])
for name in names:
  pprint.pprint(name)
```

```python
names=collection.find({'Name':'Paul Bako'})
for name in names:
  pprint.pprint(name)
```

```python
# avg_weight by each team
names=collection.aggregate([{'$group':{'_id':'$Team','avgweight':{'$avg':'$Body - Weight(lbs)'}}}])
for name in names:
  pprint.pprint(name)
```

```python
# Team who have Heaviest Player
names=collection.aggregate([{'$group':{'_id':'$Team','max_avgweight':{'$avg':'$Body - Weight(lbs)'}}},{'$sort':{'avgweight':-1}},{'$limit':1}])
for name in names:
  pprint.pprint(name)
```

```python
# all record of Maximum weight Player
names=collection.aggregate([{'$sort':{'Body - Weight(lbs)':-1}},{'$limit':1}])
for name in names:
  pprint.pprint(name)
```

```python
# Team who have Heaviest Player
names=collection.aggregate([{'$group':{'_id':'$Team','avgweight':{'$max':'$Body - Weight(lbs)'}}},{'$sort':{'avgweight':-1}},{'$limit':1}])
for name in names:
  pprint.pprint(name)
```
