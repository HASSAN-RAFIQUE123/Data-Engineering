```python
# print  resturent who have maximum (sum of score)
import pprint
# result=collection.aggregate([{'$sum':{'grades.score':-1}},{'$limit':1}])
result=collection.aggregate([{'$group':{'_id':'$name','Max':{'$sum':'$grades.score'}}},{'$sort':{'grades.score':-1}},{'$limit':5}])
for results in result:
  pprint.pprint(results)
```
