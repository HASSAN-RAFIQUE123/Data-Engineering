```python
from matplotlib import pyplot as plt
```

```python
import pandas as pd
```

```python
df=pd.read_csv("/content/company_sales_data.csv")
df
```

```python
x=df['month_number']
y=df['total_profit']
plt.plot(x,y,color="green",)
plt.xlabel("Month_Number")
plt.ylabel("Total_Profit")
plt.show()
```

```python
# Line Style dotted and Line-color should be red
# Show legend at the lower right location.
# X label name = Month Number
# Y label name = Sold units number
# Add a circle marker.
# Line marker color as read
# # Line width should be 3
x=df['month_number']
y=df['total_profit']
plt.plot(x,y,color="red",marker='o',linestyle='--',linewidth=3,markerfacecolor='green')
plt.legend(y)
plt.xlabel("Month_Number")
plt.ylabel("Total_Profit")
plt.show()
```

```python
# Read all product sales data and show it  using a multiline plot
```

```python
x=df['month_number']
y=df['facecream']
y1=df['facewash']
y2=df['bathingsoap']
y3=df['moisturizer']
y4=df['toothpaste']
plt.plot(x,y,color="red",linewidth=2,marker='o')
plt.plot(x,y2,color="blue",linewidth=2,marker='o')
plt.plot(x,y3,color="pink",linewidth=2,marker='o')
plt.plot(x,y4,color="black",linewidth=2,marker='o')
plt.plot(x,y1,color="green",linewidth=2,marker='o')
plt.legend(y)
plt.xlabel("Month_Number")
plt.ylabel("Total_Profit")
plt.show()
```

```python
#  Read toothpaste sales data of each month and show it using a scatter plot
x=df['month_number']
y=df['toothpaste']
plt.scatter(x,y)
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
# Read face cream and facewash product sales data and show it using the bar chart
fcream=df['facecream'].tolist()
fwash=df['facewash'].tolist()
month=df['month_number'].tolist()
plt.bar(month,fcream,color='red',width =0.25,label='face cream sale data')
plt.bar(month,fwash,color='green',width=0.25,label='face wash sale data ')
plt.title("Sales data")
plt.legend(loc='upper left')
plt.show()
```

```python
soap=df['bathingsoap']
month=df['month_number']
plt.bar(month,soap,width=0.25)
plt.xlabel('Months')
plt.ylabel('sales')
plt.show()
```

```python
profit=df['total_profit']
month=df['month_number']
plt.hist(profit,bins=30)
plt.xlabel('Months')
plt.ylabel('sales')
plt.show()
```
