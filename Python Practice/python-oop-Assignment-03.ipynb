```python
# class operator overloading.
class complex:
    def __init__(self, a, b):
        self.a = a
        self.b = b
# adding two objects
    def __add__(self, other): #its operators overloading
        return self.a + other.a,self.b + other.b
    def __pow__(self,m):
        return self.a**m.a,self.b**m.b
 
Ob1 = complex(1,4)
Ob2 = complex(2,3)
result = Ob1 + Ob2
result1 = Ob1 ** Ob2
print(result)
print(result1)
```
